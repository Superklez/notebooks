```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import time
```

```python
mm = pd.read_csv('make_model.csv')
mm.head()
```

```python
pbm = pd.read_csv('parent_brand_nameplate.csv')
pbm.head()
```

```python
columns = ['Make','Model','Sales Parent','Sales Brand','Sales Nameplate']
df = pd.DataFrame(columns=columns)

start_time = time.time()

for i in range(len(mm)):
    make = mm['Make'].iloc[i]
    model = mm['Model'].iloc[i]
    series_mask = pbm['Sales Nameplate'].map(lambda row: True if model in row.split() else False)
    if series_mask.sum() == 0:
        df = df.append(pd.DataFrame([[make,model,'','','']], columns=columns))
    else:
        df = df.append(pbm[(pbm['Sales Brand']==make)&(series_mask)], ignore_index=True)
        df['Make'].fillna(make, inplace=True); df['Model'].fillna(model, inplace=True)
    
time_elapsed = time.time() - start_time
print(f'Training complete in {time_elapsed % 60:.2f}s')
```
